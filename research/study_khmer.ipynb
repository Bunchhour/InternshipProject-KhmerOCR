```python
# Copy this exactly
text = "ខ្ញុំ" 

print(f"Visual Text: {text}")
print(f"Length of string: {len(text)}") # Spoiler: It is NOT 1 or 2.
print(type(text))
print("\n--- Character Breakdown ---")
for i, char in enumerate(text):
    print(f"Position {i}: {char} (Unicode: {ord(char)})")
    if char == "\u17D2":
        print("   >>> WARNING: This is the COENG sign! (Invisible Subscript Marker)")

print("\nLESSON: Your OCR must predict the COENG sign even though it is invisible!")
```
